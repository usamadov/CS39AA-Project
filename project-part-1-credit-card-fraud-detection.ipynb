# Credit Card Fraud Detection

The prevalence of credit card fraud has surged alongside the growth of digital transactions, presenting significant challenges to financial security and consumer trust. This project addresses the urgent need for effective fraud detection by harnessing advanced machine learning techniques to analyze transaction data and identify fraudulent activity. By aiming to increase detection accuracy and reduce false positives, the project seeks to bolster financial security measures, protect sensitive information, and contribute to the evolution of financial technology, ensuring the integrity and reliability of credit card transactions in the digital age.

## 1. About the Data 
The dataset includes credit card transactions from European cardholders in September 2013, characterized by a severe class imbalance where only 0.172% of the 284,807 transactions are fraudulent. Data features have been transformed using Principal Component Analysis (PCA), a statistical technique that reduces dimensionality by converting a set of correlated variables into a smaller number of uncorrelated variables called principal components, except for 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. Feature 'Class' is the response variable and it takes value 0 in case of Not Fraud and 1 in case of Fraud.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram with Time on the y-axis and Amount on the x-axis
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Time', y='Amount', bins=50, cbar=True)
plt.title('Histogram of Time vs. Amount')
plt.xlabel('Time (sec)')
plt.ylabel('Amount $')
plt.show()
```

The histogram shows the distribution of transaction amounts (y-axis) over time (x-axis) in the dataset, providing insights into how transaction amounts in dollars vary at different points in time.

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in df.columns[:-1]:  # Excluding the 'Class' column.
    plt.figure(figsize=(10, 4))
    
    # Plotting the distribution of non-fraud transactions
    sns.histplot(df[df['Class'] == 0][column], bins=50, color='blue', alpha=0.5, label='Not Fraud')
    
    # Plotting the distribution of fraud transactions
    sns.histplot(df[df['Class'] == 1][column], bins=50, color='red', alpha=1, label='Fraud')
    
    plt.title(f'Histogram for {column}')
    plt.legend()
    plt.show()
    


```

Each histogram displays the distribution of values for both non-fraudulent transactions (in blue) and fraudulent transactions (in red). These histograms provide a visual comparison of how the feature's distribution differs between the two classes (fraud and non-fraud), helping to identify potential patterns or differences in the data for fraud detection or analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming this is how you've generated your data
np.random.seed(0)
total_data_points = 284807  # Total number of data points

scater_plot = pd.DataFrame({
    'Time': np.linspace(0, 172800, total_data_points),
    'Amount': np.random.exponential(scale=200, size=total_data_points),
    'Class': np.random.choice([0, 1], size=total_data_points, p=[0.99, 0.01])
})

# After this, you can create your histogram
bins = np.linspace(0, max(scater_plot['Amount']), 30)

# Now proceed with the histogram plotting as before
plt.figure(figsize=(20, 12))
plt.hist(scater_plot[scater_plot['Class'] == 0]['Amount'], bins=bins, alpha=0.5, label='Not Fraud', color='blue')
plt.hist(scater_plot[scater_plot['Class'] == 1]['Amount'], bins=bins, alpha=1, label='Fraud', color='red')
plt.legend()
plt.title('Transaction Amounts by Class')
plt.xlabel('Amount ($)')
plt.ylabel('Number of Transactions')
plt.show()
```

The histogram shows that non-fraudulent transactions vastly outnumber fraudulent ones in the dataset. The much larger blue bars for 'Not Fraud' compared to the red 'Fraud' bars visually illustrate the imbalance between the two classes. There's a steep drop-off as transaction amounts increase, indicating that high-value transactions are rare for both classes. The imbalance, with far more examples of not fraud than fraud, is an important characteristic of the data that must be considered when analyzing and modeling the transactions.
